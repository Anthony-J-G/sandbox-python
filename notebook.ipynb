```python
number_of_columns = 10

rows = [] # We can initialize an empty list and append to it
for i in range(100):
    rows.append(normal(size=number_of_columns))


column_names = [f"Column Number {i+1}" for i in range(number_of_columns)] # or we can construct a list via an iterator like this

df = DataFrame(rows, columns=column_names)

df.head()
```
